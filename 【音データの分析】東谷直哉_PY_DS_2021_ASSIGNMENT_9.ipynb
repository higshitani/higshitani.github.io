【回答】「まず、横軸を周波数、縦軸を振幅の2乗としたときのFTTの結果に関する図を以下に示す。図1より、赤で示した部分の範囲において最も高いピークを得ていることが分かる。その範囲を拡大した様子を、次の図2に示す。

```python
# coding: utf-8
import numpy as np
from scipy.io import wavfile
import matplotlib.pyplot as plt

data = wavfile.read('voice.wav')
X = data[1]
N = len(X)
y = np.fft.fft(X)
framerate = data[0]
freq = np.linspace(0, framerate, N)
amp2 = np.absolute(y*2/N)**2

plt.figure(figsize=(27, 10)) #初期値は(6.4, 4.8)
plt.plot(freq[:N//2], amp2[:N//2], color=(0.0, 1.0, 0.0), linewidth=1.0)
plt.plot(freq[N//330:N//290], amp2[N//330:N//290], color=(1.0, 0.0, 0.0), linewidth=1.0)
plt.title("fig.1",)
plt.xlabel('FREQ [Hz]')
plt.ylabel('AMPLITUDE^2')
plt.grid(True)

plt.show()
```

図2より138～145[Hz]において高い振幅を得ていることが分かる。ゆえに、音データ(voice.wav)の音程はC#3～D3である。」

```python
# coding: utf-8
import numpy as np
from scipy.io import wavfile
import matplotlib.pyplot as plt

data = wavfile.read('voice.wav')
X = data[1]
N = len(X)
y = np.fft.fft(X)
framerate = data[0]
freq = np.linspace(0, framerate, N)
amp2 = np.absolute(y*2/N)**2

plt.figure(figsize=(27, 10)) #初期値は(6.4, 4.8)
plt.plot(freq[N//330:N//290], amp2[N//330:N//290], color=(1.0, 0.0, 0.0), linewidth=1.0)
plt.title("fig.2")
plt.xlabel('FREQ [Hz]')
plt.ylabel('AMPLITUDE^2')
plt.grid(True)

plt.show()
```
