食品成分表の分析

# 方法  
雛形を用いた、変更点は各項目ごとのcontributionを上位8個に限って表示させた。  
# 結果  
データファイルの項目順に固有値が表示された。1つ目の項目であるエネルギーから１２項目目の利用可能炭水化物の合計が0.679と全体の寄与の3分の2以上を占めた。  
# 考察
#### ①三大栄養素の妥当性について  
主成分分析の結果より、エネルギーと水分を除いたとき上位１２項目がタンパク質（アミノ酸組成によるタンパク質）・脂質（トリアシルグリセロール・飽和脂肪酸・一価不飽和脂肪酸・多価不飽和脂肪酸・コレステロール）・炭水化物（利用可能炭水化物）である。また、全５０成分のうち上記の三大栄養素とその関連成分のみで全体の寄与の３割以上を占める。エネルギーと水分を除くと、全体の半分以上の寄与率である。ゆえに、三大栄養素は妥当であると考える。  
#### ②分類について  
3大栄養素の他に、５大栄養素、さらに食物繊維を加えた６大栄養素と呼ばれる分類が存在する。これらは主成分分析において、上位から並べた時に連続して算出された。順に、タンパク質、脂質、炭水化物、食物繊維、ミネラル、ビタミンで寄与率はそれぞれ約14,24,5,6,15,7[%]である。よって、脂質を除く５つが5～15％であり、その化学的な構成から考えても、妥当性のある分類であると考えた。

```python
# coding: utf-8
import pandas as pd
import numpy as np

data = pd.read_csv('food-table.csv')
# print(data.dtypes)

cols=('ENERC','WATER','PROTAIN','PROTCAA','FAT','FATNLEA','FASAT','FAMS','FAPU','CHOLE','CH','CHOAVLM',
      'FIBSOL','FIBINS','FIBTG','ASH','NA','K','CA','MG','P','FE','ZN','CU','MN','ID','SE','CR','MO','RETOL',
      'CARTA' ,'CARTB' ,'CRYPXB','CARTBEQ','VITA_RAE','VITD','TOCPHA','TOCPHB','TOCPHG','TOCPHD','VITK','THIAHCL',
      'RIBF','NIA','VITB6A','VITB12','FOL','PANTAC','BIOT','VITC')

X = np.array(pd.DataFrame(data,columns=cols),dtype='float').T
N = X.shape[1]
M = X.shape[0]

mean = np.mean(X,axis=1)
stdev = np.std(X,axis=1)
for i in range(M):  # 標準化
    X[i,:] = (X[i,:]-mean[i])/stdev[i]

C = X.dot(X.T) / N
lmda,P = np.linalg.eig(C)
print(lmda)

print("Contribution:")
for k in range(M):
    print(k,lmda[k]/np.sum(lmda))

print("==============")
for k in range(M):
    print("Loadings for component",k)
    p2 = np.sum( P[k,:]**2 )
    loadings=[]
    for ell in range(M):
        loading = np.sqrt(lmda[k])*P[ell,k] / np.sqrt(C[ell,ell] * p2 )
        loadings.append([ell,cols[ell],loading])
    loadings_sorted = sorted(loadings, key=lambda x: abs(x[2]), reverse=True)
    for ell,name,loading in loadings_sorted[:8]:
        print(ell,name,loading)
    print("--------------")
```
